<h1 style='font-size:35px;color:#3378FF;text-aign:left'><strong>Credit Card Customer Churn Predictive Model</strong></h1>
    
Due to the rising attrition rates for our credit card customer base, we will be training a predictive model to accurately (90%+) predict if a current or new customer will eventually forgo our services. We can use the model to help us make data-driven decisions in regards to improving our services and incentives in an effort to maintain strong customer retention rates.

```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

import statsmodels.api as sm
import random

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from imblearn.over_sampling import SMOTE
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
# import dataset as pandas dataframe
df = pd.read_csv(r"...BankChurners.csv")

# view first 5 rows of dataset
df.head()
```

# Data Cleaning

```python
# view number of rows and columns
df.shape
```

The final 2 columns and CLIENTNUM will not be necessary for our analysis and predictive model. Thus, we will delete these columns before moving forward.

```python
# drop last 2 columns and CLIENTNUM
df = df.iloc[:, 0:-2]
df.drop('CLIENTNUM', axis = 1, inplace = True)

# view first 5 rows again to ensure columns have been dropped properly
df.head()
```

```python
# find out how many null values there are per column
null_counts = df.isnull().sum()
null_counts
```

There are no null values present.

```python
# drop duplicates
df.drop_duplicates(keep='first', inplace = True)

# view number of rows and columns again
df.shape
```

No duplicate values were found.

```python
# changing Attrition_Flag column to 0 if 'Existing Customer', 1 if 'Attrited Customer'
df["Attrition_Flag"] = df["Attrition_Flag"].map({"Existing Customer":0, "Attrited Customer":1})

# view first 5 rows
df.head()
```

# Exploratory Data Analysis (EDA)

```python
# viewing descriptive statistics of numerical columns
df.describe()
```

```python
# viewing information on columns of the dataset
df.info()
```

```python
# storing numerical and categorical columns in separate dataframes
cat_cols = list(df.select_dtypes("object"))
num_cols = list(df.select_dtypes(["int64", "float64"]))
```

## Categorical Columns

```python
fig, ax = plt.subplots(ncols=1, nrows=5, figsize=(14, 26))

# counter
i = 0

# for loop to iterate through categorical columns, create bar plots displaying rate of churn
for cols in cat_cols:
    sns.barplot(x=df[cols], y=df['Attrition_Flag'], fill=True, alpha=1, errorbar=None, ax=ax[i], palette=('#3378FF', '#9EBDFF', '#E9DAC1'))
    ax[i].set_xlabel(' ')
    ax[i].set_xlabel(' ')
    ax[i].set_ylabel(' ')
    ax[i].xaxis.set_tick_params(labelsize=14)
    ax[i].tick_params(left=False, labelleft=False)
    ax[i].set_ylabel(cols, fontsize=16)    
    ax[i].bar_label(ax[i].containers[0], size="12")
    ax[i].grid(False)
    i=i+1

# display bar plots
plt.show()
```

## Numerical Columns

```python
fig, ax = plt.subplots(ncols=2, nrows=14, figsize=(14, 40))

for i, col in enumerate(num_cols[1:]):
    # removing outliers
    Q3 = df[col].quantile(0.99)
    df_filtered = df[df[col] <= Q3]
    Q1 = df[col].quantile(0.01)
    df_filtered = df_filtered[df_filtered[col] >= Q1]
    
    sns.kdeplot(data=df_filtered, x=col, fill=True, alpha=1, hue='Attrition_Flag',
            palette=('#3378FF', '#E9DAC1'), multiple='stack', ax=ax[i, 0], warn_singular=False)
    sns.boxplot(data=df_filtered, y=col, x='Attrition_Flag', ax=ax[i, 1],
                palette=('#3378FF', '#E9DAC1'), color='white')
    
    ax[i, 0].set_xlabel(' ')
    ax[i, 1].set_xlabel(' ')
    ax[i, 1].set_ylabel(' ')
    ax[i, 1].xaxis.set_tick_params(labelsize=14)
    ax[i, 0].tick_params(left=False, labelleft=False)
    ax[i, 0].set_ylabel(col, fontsize=12)

plt.show()
```

## Correlation for Numerical Features
The correlation heatmap below displays all our numerical feature columns and their correlation coefficients with our response variable, Attrition_Flag.

```python
# correlation coefficient heatmap for numerical feature columns and Attrition_Flag
selected_data = df[num_cols]

correlation_matrix = selected_data.corr(numeric_only=True)

plt.figure(figsize=(18,8))
sns.heatmap(correlation_matrix, annot=True, cmap='Blues', linewidths=0.5)
plt.title('Correlation for Numerical Features', size=25, color = '#3378FF')
plt.show()
```

## Existing and Churned Customer Proportions

The pie chart below displays the inequality in proportions of customers that are still current customers of the bank, and those that have forgone our services. Only approximately 16.1% of our credit card customer base have churned.

```python
# creating a pie chart displaying the proportion of customers that are existing customers and customers that are attrited
existing_count = df['Attrition_Flag'].value_counts()[0]
churned_count = df['Attrition_Flag'].value_counts()[1]

labels = ['Existing Customers', 'Churned Customers']

counts = [existing_count, churned_count]

colors = ['#3378FF', '#E9DAC1']

plt.figure(figsize=(6, 6))
plt.pie(counts, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90, explode=(0, 0.05))
plt.title('Proportion of Existing and Churned Customers')

plt.show()
```

# Data Prep for Model Training

## Encoding categorical variables

```python
for col in cat_cols:
    dummy_cols = pd.get_dummies(df[col], drop_first=True, prefix=col)
    df = pd.concat([df,dummy_cols],axis=1)
    df.drop(columns=col, inplace=True)
```

## Train-test split

```python
# splitting data into training and testing folds
X = df
y = df.pop('Attrition_Flag')

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
feature_cols = ['Customer_Age', 'Dependent_count', 
                    'Months_on_book', 'Total_Relationship_Count', 'Months_Inactive_12_mon',
                    'Contacts_Count_12_mon', 'Credit_Limit', 'Total_Revolving_Bal',
                    'Avg_Open_To_Buy', 'Total_Amt_Chng_Q4_Q1', 'Total_Trans_Amt', 'Total_Trans_Ct',
                    'Total_Ct_Chng_Q4_Q1', 'Avg_Utilization_Ratio']

# using MinMaxScaler() to normalize data to a common scale
scaler = MinMaxScaler()

X_train = scaler.fit_transform(X_train[feature_cols])

X_test = scaler.transform(X_test[feature_cols])
```

The pie chart above displays the significant imbalance between existing and churned customers. To address this issue, we will be applying SMOTE to the training data.

```python
# applying SMOTE to address imbalance in data
smote = SMOTE(random_state=42, sampling_strategy=0.839)
X_train, y_train = smote.fit_resample(X_train, y_train)
```

# Building the Predictive Model

## Random Forest Classifier

```python
# instantiating random forest algorithm
rf = RandomForestClassifier(random_state=42)

rf.fit(X_train, y_train)

y_pred = rf.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print('Accuracy: ', accuracy)
```

## Classification Report

```python
# retrieving classification report
print(classification_report(y_test, y_pred))
```

## Confusion Matrix

```python
# displaying confusion matrix as sns heatmap
conf_matrix = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(10, 5))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', square=True, 
            xticklabels=['Existing', 'Churned'],
            yticklabels=['Existing', 'Churned'])

plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('Confusion Matrix')

plt.show()
```

## Feature Importance

```python
feature_importance = rf.feature_importances_
```

```python
# feature importance of numerical columns
feature_importance_df = pd.DataFrame({'Feature': feature_cols, 'Importance': feature_importance})

feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x='Importance', y='Feature', data=feature_importance_df, palette='Blues_r')

plt.xlabel('Feature Importance')
plt.ylabel('Feature')
plt.title('Feature Importance in Random Forest Classifier')

plt.show()
```
